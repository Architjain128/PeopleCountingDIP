```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
from queue import Queue
```

```python
def bfs(img,r,c,col,vis):
    h,w = img.shape
    q = Queue()
    q.put((r,c))
    vis[r,c] = 1
    cnt = 0
    while not q.empty():
        cnt += 1
        r, c = q.get()
        if r-1>=0 and vis[r-1,c] == 0 and img[r-1,c]!=0:
            vis[r-1,c] = col
            q.put((r-1,c))
        if r+1<h and vis[r+1,c] == 0 and img[r+1,c]!=0:
            vis[r+1,c] = 1
            q.put((r+1,c))
        if c-1>=0 and vis[r,c-1] == 0 and img[r,c-1]!=0:
            vis[r,c-1] = 1
            q.put((r,c-1))
        if c+1<w and vis[r,c+1] == 0 and img[r,c+1]!=0:
            vis[r,c+1] = 1
            q.put((r,c+1))
            
    return cnt
```

```python
img = cv2.imread('img.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
plt.title('original image')
plt.xticks([]), plt.yticks([]) 
plt.show()

# taking OSTU's threshold for best results
th, thresh1 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

img = (img > th).astype(np.uint8)
plt.imshow(img, cmap='gray')
plt.title('image thresholding')
plt.xticks([]), plt.yticks([]) 
plt.show()

img = cv2.dilate(img, cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3)), iterations = 3)
plt.imshow(img, cmap='gray')
plt.title('dilation')
plt.xticks([]), plt.yticks([]) 
plt.show()
```

```python
vis = np.zeros(img.shape)
h,w = img.shape
arr = []
sum = 0
for i in range(h):
    for j in range(w):
        if vis[i,j]==0 and img[i,j]==1:
            cnt = bfs(img,i,j,1,vis)
            sum += cnt
            arr.append((i,j,cnt))
        
# taking threshold as the average of all the counts 
print('size of connected components are: ',arr[0][2],arr[1][2],arr[2][2])
l = len(arr)
# this k will determine if, the connected component will be considered background or not?
k = sum//l
print(k)
```

```python
fin_arr = []
for i in range(l):
    if arr[i][2] >= k:
        fin_arr.append((arr[i][0],arr[i][1]))
        
vis = np.zeros(img.shape)
c = 1
for i in range(len(fin_arr)):
    cnt = bfs(img,fin_arr[i][0],fin_arr[i][1],c,vis)
    c += 1
    
print('number of regions of interest are: ',len(fin_arr))
```

```python
COLORS = [(255, 0, 0), (0, 255, 0), (0,0,255), (255,255,0), (255,0,255), (0,255,255)]
final_img = np.zeros((h, w,3), int)
# finding centroid
cent_x = 0.0
cent_y = 0.0
cnt = 0
for i in range(h):
    for j in range(w):
        if vis[i,j] != 0:
            col = math.floor(vis[i,j]%6)   #we have 6 colors in above array
            final_img[i,j,0] = COLORS[col][0] 
            final_img[i,j,1] = COLORS[col][1] 
            final_img[i,j,2] = COLORS[col][2] 
            
            cnt += 1
            cent_x += i
            cent_y += j
            
plt.imshow(final_img)
plt.title('region of interset: colored')
plt.xticks([]), plt.yticks([]) 
plt.show()

# we need integer centroid co-ordinates
cent_x = cent_x//cnt
cent_y = cent_y//cnt

print('Centroid of image is at: ',cent_x, cent_y)
```
